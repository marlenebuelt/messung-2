```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_bl=pd.read_csv('/Users/marlenebultemann/Desktop/HTW/Bachelorarbeit/messungen_64bit_results/bl_messung/df_bl')
df_dt=pd.read_csv('/Users/marlenebultemann/Desktop/HTW/Bachelorarbeit/messungen_64bit_results/1_dt_results/df_dt')
df_lr=pd.read_csv('/Users/marlenebultemann/Desktop/HTW/Bachelorarbeit/messungen_64bit_results/2_logreg/df_lr')
df_svm=pd.read_csv('/Users/marlenebultemann/Desktop/HTW/Bachelorarbeit/messungen_64bit_results/3_svm/df_svm')
df_nn=pd.read_csv('/Users/marlenebultemann/Desktop/HTW/Bachelorarbeit/messungen_64bit_results/04_nn_results/df_nn')
```

```python
cpu_bl = df_bl['CPU Utilization']
mem_bl = df_bl['%mem']
power_total_bl = df_bl['Total Power']

cpu_dt = df_dt['CPU Utilization']
mem_dt = df_dt['%mem']
power_total_dt = df_dt['Total Power']

cpu_lr = df_lr['CPU Utilization']
mem_lr = df_lr['%mem']
power_total_lr = df_lr['Total Power']

cpu_svm = df_svm['CPU Utilization']
mem_svm = df_svm['%mem']
power_total_svm = df_svm['Total Power']

cpu_nn = df_nn['CPU Utilization']
mem_nn = df_nn['%mem']
power_total_nn = df_nn['Total Power']
```

```python
# Boxplot labels
labels = ['BL', 'DT', 'LR', 'SVM', 'MLP']

cpu_data = [cpu_bl, cpu_dt, cpu_lr, cpu_svm, cpu_nn]
mem_data = [mem_bl, mem_dt, mem_lr, mem_svm, mem_nn]
power_data = [power_total_bl, power_total_dt, power_total_lr, power_total_svm, power_total_nn]

# Create three boxplots for CPU, RAM, and Power
plt.figure(figsize=(12, 6))

# Boxplot for CPU Utilization
plt.subplot(1, 3, 1)
plt.boxplot(cpu_data, labels=labels)
plt.title('CPU Utilization')
plt.ylabel('% CPU')

# Boxplot for Memory
plt.subplot(1, 3, 2)
plt.boxplot(mem_data, labels=labels)
plt.title('Memory Usage')
plt.ylabel('% Memory')

# Boxplot for Power
plt.subplot(1, 3, 3)
plt.boxplot(power_data, labels=labels)
plt.title('Total Power')
plt.ylabel('Power')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
dt_ws = ((np.cumsum(power_total_dt).iloc[-1])/30)*inferences_year
lr_ws = ((np.cumsum(power_total_lr).iloc[-1])/30)*inferences_year
svm_ws = ((np.cumsum(power_total_svm).iloc[-1])/30)*inferences_year
nn_ws = ((np.cumsum(power_total_nn).iloc[-1])/30)*inferences_year

print(dt_ws)
print(lr_ws)
print(svm_ws)
print(nn_ws)
```

```python
dt_wh = dt_ws/3600
lr_wh = lr_ws/3600
svm_wh = svm_ws/3600
nn_wh = nn_ws/3600

print(dt_wh)
print(lr_wh)
print(svm_wh)
print(nn_wh)
```

```python
price_kwh = 0.3496
def price_total(model_wh):
    kwh = model_wh/1000
    result = kwh*price_kwh
    return result
```

```python
print(price_total(dt_wh))
print(price_total(lr_wh))
print(price_total(svm_wh))
print(price_total(nn_wh))
```
